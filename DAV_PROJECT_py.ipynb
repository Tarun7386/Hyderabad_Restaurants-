```python
def convert(price):
    price = price / 74  # Convert to USD
    price = price + 1.28  # Add additional cost
    return '$ {:.2f}'.format(price)  # Format the price as a string with 2 decimal places and prepend '$'
```

```python
df['Price_in_USD'] = df['price for one'].apply(lambda x: convert(x))
df.head()
```
